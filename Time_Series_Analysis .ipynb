# Splitting data

```python
Train = df.head(6070)
Test = df.tail(1518)
```

```python
Train
```

```python
Test
```

```python
from sklearn.metrics import mean_squared_error
```

```python
#Linear Model
import statsmodels.formula.api as smf 

linear_model = smf.ols('Ex_rate~t',data=Train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(mean_squared_error(Test['Ex_rate'], pred_linear))
rmse_linear
```

# While looking at that mentioned accuracy, it is almost equivalent to zero, indicating that we have constructed a good model..

# 
**Null hypothesis: Data is Non stationary**

**Alternate Hypothesis: Data is Stationary**
   


- If p< 0.05 ; Data is stationary

- if p>0.05; Data is not stationary

**According the p-value, the dataset is Non-stationary as the p value is more than alpha value.**
